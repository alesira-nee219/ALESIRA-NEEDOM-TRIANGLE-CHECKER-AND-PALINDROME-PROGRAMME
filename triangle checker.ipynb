```python
# Hardcoded angles - no input() needed here, we already know the values.
angle1 = 30
angle2 = 60
angle3 = 90

print("=== Right-Angled Triangle Checker (Example) ===\n")

# Confirm the three angles add up to a valid triangle (should always be 180).
total = angle1 + angle2 + angle3
print(f"Angles: {angle1}°, {angle2}°, {angle3}°  (sum = {total}°)")

angles = [angle1, angle2, angle3]

# A small tolerance handles rounding issues for decimal angles (not needed
# here since these are whole numbers, but keeps the logic consistent).
tolerance = 1e-6
is_right_angled = any(abs(a - 90) < tolerance for a in angles)

if is_right_angled:
    print("Result: This IS a right-angled triangle. ✅")
else:
    print("Result: This is NOT a right-angled triangle. ❌")
```

```python

"""
121 reversed is still 121, so this number IS a palindrome.
"""

# Hardcoded number - no input() needed here, we already know the value.
number = 121

print("=== Palindrome Number Checker ===\n")
print(f"Checking number: {number}")

original = number
reversed_number = 0
n = number

#rebuild the number in reverse.
while n > 0:
    last_digit = n % 10          # remainder -> isolates the last digit
    reversed_number = (reversed_number * 10) + last_digit
    n = n // 10                  # whole quotient -> chops off that digit

print(f"Reversed number: {reversed_number}")

if original == reversed_number:
    print(f"Result: {number} IS a palindrome. ✅")
else:
    print(f"Result: {number} is NOT a palindrome. ❌")
```
